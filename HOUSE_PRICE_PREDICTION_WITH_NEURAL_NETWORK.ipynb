# Create a model and train the model 

```python
X = df.drop('price', axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras import models, layers
```

```python
Xtrain, Xtest, ytrain ,ytest = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
scaler =MinMaxScaler()
```

```python
scaler.fit(Xtrain)
```

```python
Xtrain = scaler.transform(Xtrain)
Xtest = scaler.transform(Xtest)
```

```python
network_0 = models.Sequential()
network_0.add(layers.Dense(19, activation='relu'))
network_0.add(layers.Dense(19, activation='relu'))
network_0.add(layers.Dense(19, activation='relu'))
network_0.add(layers.Dense(19, activation='relu'))

network_0.add(layers.Dense(1))
network_0.compile(optimizer='adam', 
               loss='mse')
```

```python
network_0.fit(Xtrain, ytrain, epochs=100, batch_size=128,validation_data=(Xtest, ytest), verbose=1)
```

```python
history_0 = pd.DataFrame(network_0.history.history)
```

```python
history_0.plot() # seems like we could continue train the data since both lines are dropping 
```

```python
network = models.Sequential()
network.add(layers.Dense(19, activation='relu'))
network.add(layers.Dense(19, activation='relu'))
network.add(layers.Dense(19, activation='relu'))
network.add(layers.Dense(19, activation='relu'))
network.add(layers.Dense(1))
network.compile(optimizer='adam', 
               loss='mse')
```

```python
network.fit(Xtrain, ytrain, epochs=400, batch_size=128,validation_data=(Xtest, ytest), verbose=50)
```

```python
history = pd.DataFrame(network.history.history)
```

```python
history.plot(); # although we can still keep on, this looks good enough, we'll stop here
```

# Model valuation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
pred = network.predict(Xtest)
```

```python
mean_squared_error(ytest, pred)**0.5
```

```python
mean_absolute_error(ytest, pred)
```

```python
df['price'].describe() # our mean is much lower, oh no
```

```python
explained_variance_score(ytest, pred) # our model explains about 76% of the variance, remember we could still continue our training, since we are not overfitting the data yet
```

```python
plt.figure(figsize=(10,5))
plt.scatter(ytest, pred)
plt.plot(ytest, ytest, 'r-'); # not too bad
```
